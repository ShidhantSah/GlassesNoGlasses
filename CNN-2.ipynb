```python
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator # type: ignore
```

```python
test_df = pd.read_csv('./Kaggle-Datasets/glassesnoglasses/test.csv')
dir = './Kaggle-Datasets/glassesnoglasses/faces-spring-2020/face-'
test_df['path'] = test_df['id'].apply(lambda x: dir + str(x) + '.png').tolist()

display(test_df)
```

```python
data = test_df.dropna()
```

```python
test_datagen = ImageDataGenerator(rescale = 1.0/255)


test_gen = test_datagen.flow_from_dataframe(
    dataframe=data,
    x_col='path',
    target_size=(150, 150),
    batch_size=32,
    class_mode=None,
)
```

```python
from tensorflow.keras.models import load_model # type: ignore

model = load_model('./GorNG.h5')
```

```python
loss, accuracy = model.evaluate(test_gen)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```
